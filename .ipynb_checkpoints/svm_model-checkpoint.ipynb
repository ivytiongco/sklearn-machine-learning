```python
# SVM - Support Vector Machine 

import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame and Perform Basic Data Cleaning

df = pd.read_csv('exoplanet_data.csv')
df

# koi_disposition: CANDIDATE, CONFIRMED, FALSE POSITIVE
# koi_pdisposition: CANDIDATE, FALSE POSITIVE
```

```python
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
#Create a Train Test Split
# Use koi_disposition for the y values

from sklearn.model_selection import train_test_split
y = df["koi_disposition"]
X = df.drop(columns=["koi_disposition"])

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
# Pre-processing
# Scale the data using the MinMaxScaler

from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Train the Model

# Support vector machine linear classifier
from sklearn.svm import SVC 
model2 = SVC(kernel='linear')
model2.fit(X_train_scaled, y_train)
```

```python
# Hyperparameter Tuning
# Use GridSearchCV to tune model's parameters

from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 11],
              'gamma': [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model2, param_grid, verbose=3)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
# save the fitted model to file
import joblib
filename = 'svm.sav'
joblib.dump(grid, filename)
```
